```python
df_dummies=pd.get_dummies(data=df2, columns=['weight_class','B_Stance','R_Stance','no_of_rounds'], drop_first=True)
df_dummies.head()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
sns.scatterplot(x="R_Height_cms", y="R_Weight_lbs", color = 'blue', data=df2, ax = ax, size="size")
sns.scatterplot(x="B_Height_cms", y="B_Weight_lbs", color = 'blue', data=df2, ax = ax, size="size")
ax.set(xlabel='Weight (lbs)', ylabel='Height (cm)')
plt.title("Heights vs Weights")
# add B_Weight
# clean up x axis labels
# clean x and y axis names
```

```python
df2.sort_values(by=['weight_class'], inplace = True)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
sns.scatterplot(x="weight_class", y="R_Weight_lbs", color = 'blue', data=df2, ax = ax)
sns.scatterplot(x="weight_class", y="B_Weight_lbs", color = 'blue', data=df2, ax = ax)
ax.set(xlabel='Weight Class', ylabel='Weight (lbs)')
plt.title("Weights per Weight Class")
plt.xticks(rotation=65)
plt.show()
```

```python
info = pd.DataFrame(df2.groupby('weight_class')['R_Weight_lbs', 'B_Weight_lbs'].max().reset_index())
info['newcol'] = info.max(axis = 1)
test = df2.merge(info[['weight_class','newcol']], how = 'left', on = 'weight_class')
test.sort_values(by=['newcol'], inplace = True)
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
sns.scatterplot(x="weight_class", y="R_Weight_lbs", color = 'blue', data=test, ax = ax)
sns.scatterplot(x="weight_class", y="B_Weight_lbs", color = 'blue', data=test, ax = ax)
ax.set(xlabel='Weight Class', ylabel='Weight (lbs)')
plt.title("Weights per Weight Class")
plt.xticks(rotation=65)
plt.show()
```
